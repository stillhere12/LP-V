```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

df = pd.read_csv("vector_result.txt")
max_n = int(df["N"].max())
xticks = list(range(0, max_n + 1, max_n // 5))
```

```python
# ================= VECTOR ADD PERFORMANCE =================
plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
seq = np.maximum(df["SEQ"], 1e-6)
par = np.maximum(df["PAR"], 1e-6)
plt.plot(df["N"], seq, 'b*-', label='Sequential')
plt.plot(df["N"], par, 'g*-', label='Parallel')
plt.yscale('log')
plt.gca().yaxis.set_major_formatter(ScalarFormatter())
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Time (s)")
plt.title("Vector Addition Performance")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= VECTOR ADD SPEEDUP =================
plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df["N"], df["SPEEDUP"], 'r*-', label='Speedup')
plt.axhline(y=1, color='k', linestyle='--', alpha=0.5)
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Speedup")
plt.title("Vector Addition Speedup")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= VECTOR ADD EFFICIENCY =================
plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df["N"], df["EFFICIENCY"], 'm*-', label='Efficiency')
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Efficiency")
plt.title("Vector Addition Efficiency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= VECTOR ADD COST =================
plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df["N"], df["COST"], 'y*-', label='Cost')
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Cost (sec-cores)")
plt.title("Vector Addition Cost")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
