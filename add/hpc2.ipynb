```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter

xticks = [0, 20000, 40000, 60000, 80000, 100000]
```

```python
# ================= BUBBLE SORT PERFORMANCE =================
df = pd.read_csv("bubble_result.csv")

plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df['N'], df['SEQ'] * 1000, 'b*-', label='Sequential')
plt.plot(df['N'], df['PAR'] * 1000, 'g*-', label='Parallel')
plt.yscale('log')
plt.gca().yaxis.set_major_formatter(ScalarFormatter())
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Time (ms)")
plt.title("Bubble Sort Performance")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= BUBBLE SORT SPEEDUP =================
df = pd.read_csv("bubble_result.csv")

plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df['N'], df['SPEEDUP'], 'r*-', label='Speedup')
plt.axhline(y=1, color='k', linestyle='--', alpha=0.5)
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Speedup")
plt.title("Bubble Sort Speedup")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= BUBBLE SORT EFFICIENCY =================
df = pd.read_csv("bubble_result.csv")

plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df['N'], df['EFFICIENCY'], 'm*-', label='Efficiency')
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Efficiency")
plt.title("Bubble Sort Efficiency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= BUBBLE SORT COST =================
df = pd.read_csv("bubble_result.csv")

plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df['N'], df['COST'] * 1000, 'y*-', label='Cost')
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Cost (ms-cores)")
plt.title("Bubble Sort Cost")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= MERGE SORT PERFORMANCE =================
df = pd.read_csv("merge_result.csv")

plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')

# Replace 0 values with a small floor to avoid log(0) issues
seq_ms = np.maximum(df['SEQ'] * 1000, 1e-3)
par_ms = np.maximum(df['PAR'] * 1000, 1e-3)

plt.plot(df['N'], seq_ms, 'b*-', label='Sequential')
plt.plot(df['N'], par_ms, 'g*-', label='Parallel')
plt.yscale('log')
plt.gca().yaxis.set_major_formatter(ScalarFormatter())
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Time (ms)")
plt.title("Merge Sort Performance")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= MERGE SORT SPEEDUP =================
df = pd.read_csv("merge_result.csv")

plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df['N'], df['SPEEDUP'], 'r*-', label='Speedup')
plt.axhline(y=1, color='k', linestyle='--', alpha=0.5)
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Speedup")
plt.title("Merge Sort Speedup")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
