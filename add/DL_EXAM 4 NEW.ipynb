```python
# =========================================================
# GOOGLE STOCK PRICE PREDICTION USING LSTM AND GRU
# =========================================================

# =====================
# IMPORT LIBRARIES
# =====================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import GRU
from tensorflow.keras.layers import Dropout

# =====================
# LOAD DATASET
# =====================

df = pd.read_csv("googl_data_2020_2025.csv")

print(df.head())

# =====================
# CLEAN DATASET
# =====================

# Remove unwanted rows
df = df.iloc[2:].reset_index(drop=True)

# Rename column
df.rename(columns={'Price':'Date'}, inplace=True)

# Convert date column
df['Date'] = pd.to_datetime(df['Date'])

# Convert numeric columns
numeric_cols = ['Adj Close', 'Close',
                'High', 'Low',
                'Open', 'Volume']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col],
                            errors='coerce')

# Remove null values
df.dropna(inplace=True)

print(df.head())

# =====================
# VISUALIZE DATA
# =====================

plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Close'])

plt.title('Google Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Close Price')

plt.grid(True)
plt.show()

# =====================
# PREPARE DATA
# =====================

data = df[['Close']]

# Normalize data
scaler = MinMaxScaler(feature_range=(0,1))

scaled_data = scaler.fit_transform(data)

# =====================
# CREATE SEQUENCES
# =====================

X = []
y = []

sequence_length = 60

for i in range(sequence_length,
               len(scaled_data)):

    X.append(
        scaled_data[i-sequence_length:i, 0]
    )

    y.append(scaled_data[i, 0])

X = np.array(X)
y = np.array(y)

# Reshape for RNN
X = np.reshape(
    X,
    (X.shape[0], X.shape[1], 1)
)

print(X.shape)
print(y.shape)

# =====================
# TRAIN TEST SPLIT
# =====================

split = int(0.8 * len(X))

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]

# =========================================================
# LSTM MODEL
# =========================================================

print("\n========================")
print("TRAINING LSTM MODEL")
print("========================")

lstm_model = Sequential()

# First LSTM Layer
lstm_model.add(
    LSTM(
        units=50,
        return_sequences=True,
        input_shape=(X_train.shape[1],1)
    )
)

lstm_model.add(Dropout(0.2))

# Second LSTM Layer
lstm_model.add(
    LSTM(
        units=50,
        return_sequences=True
    )
)

lstm_model.add(Dropout(0.2))

# Third LSTM Layer
lstm_model.add(LSTM(units=50))

lstm_model.add(Dropout(0.2))

# Output Layer
lstm_model.add(Dense(units=1))

# Compile
lstm_model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)

lstm_model.summary()

# Train
lstm_history = lstm_model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)

# =====================
# LSTM PREDICTIONS
# =====================

lstm_predictions = lstm_model.predict(X_test)

lstm_predictions = scaler.inverse_transform(
    lstm_predictions
)

real_prices = scaler.inverse_transform(
    y_test.reshape(-1,1)
)

# =====================
# LSTM EVALUATION
# =====================

lstm_mse = mean_squared_error(
    real_prices,
    lstm_predictions
)

lstm_rmse = np.sqrt(lstm_mse)

lstm_mae = mean_absolute_error(
    real_prices,
    lstm_predictions
)

lstm_mape = mean_absolute_percentage_error(
    real_prices,
    lstm_predictions
)

lstm_accuracy = 100 - (lstm_mape * 100)

print("\n======= LSTM RESULTS =======")

print("LSTM RMSE :", lstm_rmse)
print("LSTM MAE :", lstm_mae)

print(
    "LSTM Accuracy : {:.2f}%"
    .format(lstm_accuracy)
)

# =========================================================
# GRU MODEL
# =========================================================

print("\n========================")
print("TRAINING GRU MODEL")
print("========================")

gru_model = Sequential()

# First GRU Layer
gru_model.add(
    GRU(
        units=50,
        return_sequences=True,
        input_shape=(X_train.shape[1],1)
    )
)

gru_model.add(Dropout(0.2))

# Second GRU Layer
gru_model.add(
    GRU(
        units=50,
        return_sequences=True
    )
)

gru_model.add(Dropout(0.2))

# Third GRU Layer
gru_model.add(GRU(units=50))

gru_model.add(Dropout(0.2))

# Output Layer
gru_model.add(Dense(units=1))

# Compile
gru_model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)

gru_model.summary()

# Train
gru_history = gru_model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)

# =====================
# GRU PREDICTIONS
# =====================

gru_predictions = gru_model.predict(X_test)

gru_predictions = scaler.inverse_transform(
    gru_predictions
)

# =====================
# GRU EVALUATION
# =====================

gru_mse = mean_squared_error(
    real_prices,
    gru_predictions
)

gru_rmse = np.sqrt(gru_mse)

gru_mae = mean_absolute_error(
    real_prices,
    gru_predictions
)

gru_mape = mean_absolute_percentage_error(
    real_prices,
    gru_predictions
)

gru_accuracy = 100 - (gru_mape * 100)

print("\n======= GRU RESULTS =======")

print("GRU RMSE :", gru_rmse)
print("GRU MAE :", gru_mae)

print(
    "GRU Accuracy : {:.2f}%"
    .format(gru_accuracy)
)

# =========================================================
# COMPARE BOTH MODELS
# =========================================================

print("\n===================================")
print("LSTM vs GRU COMPARISON")
print("===================================")

print("LSTM Accuracy : {:.2f}%"
      .format(lstm_accuracy))

print("GRU Accuracy : {:.2f}%"
      .format(gru_accuracy))

# =========================================================
# VISUALIZATION
# =========================================================

plt.figure(figsize=(14,6))

plt.plot(real_prices,
         color='blue',
         label='Actual Price')

plt.plot(lstm_predictions,
         color='red',
         label='LSTM Prediction')

plt.plot(gru_predictions,
         color='green',
         label='GRU Prediction')

plt.title(
    'Google Stock Price Prediction'
)

plt.xlabel('Time')
plt.ylabel('Stock Price')

plt.legend()
plt.grid(True)

plt.show()

# =========================================================
# FUTURE PREDICTION
# =========================================================

last_60_days = scaled_data[-60:]

X_future = []

X_future.append(last_60_days)

X_future = np.array(X_future)

X_future = np.reshape(
    X_future,
    (
        X_future.shape[0],
        X_future.shape[1],
        1
    )
)

# LSTM Future Prediction
lstm_future = lstm_model.predict(X_future)

lstm_future = scaler.inverse_transform(
    lstm_future
)

# GRU Future Prediction
gru_future = gru_model.predict(X_future)

gru_future = scaler.inverse_transform(
    gru_future
)

print("\n===================================")
print("NEXT DAY STOCK PRICE PREDICTION")
print("===================================")

print("LSTM Predicted Price :",
      lstm_future[0][0])

print("GRU Predicted Price :",
      gru_future[0][0])
```
