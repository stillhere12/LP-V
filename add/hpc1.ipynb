```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

# Read the data
df = pd.read_csv("result.txt")
xticks = [0, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000]
```

```python
# ================= BFS PERFORMANCE =================
plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df['N'], df['BFS_SEQ'] * 1000, 'b*-', label='Sequential')
plt.plot(df['N'], df['BFS_PAR'] * 1000, 'g*-', label='Parallel')
plt.yscale('log')
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:g}'))
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Time (ms)")
plt.title("BFS Sequential vs Parallel Performance")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# ================= BFS SPEEDUP =================
plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.plot(df['N'], df['BFS_SPEEDUP'], 'r*-', label='Speedup')
plt.xticks(xticks)
plt.xlabel("Input Size N")
plt.ylabel("Speedup")
plt.title("BFS Speedup")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
